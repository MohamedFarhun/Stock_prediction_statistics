```python
def list_duplicates(seq):
  seen = set()
  seen_add = seen.add
  seen_twice = set( x for x in seq if x in seen or seen_add(x) )
  return list( seen_twice )
list_duplicates(x)
```

```python
import distance
score = []
for i in range(len(x_sample)):
    x_target = x_sample[i:i+len(x)] #rolling the Pattern over windows of SampleTarget
    score.append(distance.levenshtein(x, x_target))
print(score)
```

```python
x2 = np.arange(start = np.argmin(score),stop= np.argmin(score)+len(x))
plt.figure(figsize=(16,12))
plt.plot(x0,x_sample, c='DarkBlue',linewidth=4)
plt.plot(x2,x)
```

```python
def primes_method(n):
    out = list()
    sieve = [True] * (n+1)
    for p in range(2, n+1):
        if (sieve[p]):
            out.append(p)
            for i in range(p, n+1, p):
                sieve[i] = False
    return out
```

```python
def findPrime(n): 
        flag = 0  
        if n > 1: 
            for i in np.arange(2, n//2): 
                if (n % i) == 0: 
                    flag = 0 
                    break 
                else: 
                    flag = 1 
        else: 
            flag = 0 
        return flag 
n = x
for i in range(len(n)): 
  f = findPrime(n[i]) 
  if f == 0:
    print(n[i]," is not a prime number") 
  else: 
    print(n[i]," is a prime number") 
```
