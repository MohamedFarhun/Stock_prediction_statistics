```python
import numpy as np
from numpy.linalg import norm
import warnings
warnings.filterwarnings("ignore") 
import yfinance as yf
yf.pdr_override()
```

```python
symbol = 'AMD'
start = '2021-08-20'
end = '2022-08-20'
dataset = yf.download(symbol,start,end)
dataset.tail()
```

```python
Open = np.array(dataset['Open'])
Close = np.array(dataset['Adj Close'])
```

```python
norm(Close - Open)
```
